##Number Guessing Game

######Problem Statement: Create a Python program that implements a simple number guessing game using a while loop. The program should make use of control statements such as else, break, and continue.

```python
#★ Generate a random number between 1 and 10 that the user has to guess.Import random and use randint function.
import random

## Generate random number between 1 and 10
secret_number=random.randint(1,10)
print("Secret number is",secret_number)

#★ Set a variable attempts to 3, which represents the maximum number of guesses allowed.
attempts= 3

#★ Use a while loop to allow the user to keep guessing until they get the correct number or run out of attempts.
while(attempts>0):

  #★ Ask the user to guess the number.
    guess = int(input("Guess the number (between 1 and 10): "))

  #★ Provide feedback to the user for each guess:
    if(guess<1 or guess >10):
      print("Your guess is out of range. Please guess a number between 1 and 10.")
      attempts-=1
      continue
    elif(guess>secret_number):
      print("Too high. Try again.")
      attempts-=1
    elif(guess<secret_number):
      print("Too low. Try again.")
      attempts-=1
    elif(guess==secret_number):
      print("Congratulations! You guessed the correct number.")
      break
else:
  print("Sorry!You have reached maximum number of attempts!Better luck next time!")
```

##Multiplication Table Generator

###Problem Statement: Create a Python program that generates and prints a multiplication table (from 1 to 10) for a given number using a for loop and the range function.

```python
#1. Prompt user for Input. Ask the user to enter a number for which they want to generate a multiplication table.
number = int(input("Enter the number for which you want the multiplication table: "))

#Use a for loop to iterate through the numbers 1 to 10. In each iteration, calculate the product of the user's number and the current number from the loop.
for i in range(1,11):
  print(f"{number} x {i} =",number*i)
```

##BMI Calculator

######Problem Statement: Create a Python program that calculates the Body Mass Index(BMI).

```python
#Define a function calculate_bmi(weight, height) that returns the BMI.
def calculate_bmi(weight,height):
  BMI = round(weight/ (height**2),1)
  return BMI

#Prompt the user for their weight (in kg) and height (in meters).
weight = int(input("Enter your weight in kg: " ))
height = float(input("Enter your height in m: "))

#Use the function to calculate and display the BMI
result = calculate_bmi(weight,height)
print((f"Your Body Mass Index is:{result}"))
```
