#Strings

## 1. String Concatenation


```
# Write a Python program that takes two strings i.e string 1 “Hello ”, string 2
# get name as input from the user and concatenates them together. Display the concatenated string as the output.
string_1 = "Hello"
string_2 = input('Enter your name ')
str_concat1 = string_1 + " " + string_2
print(str_concat1)



# Now concatenate string 3 “, welcome to Python programming” to the existing string and display the output string.
string_3 = "welcome to Python programming"
Str_concat=str_concat1 + ", " + string_3
print(Str_concat)
```

##2. String Slicing and Indexing


```
# Write a Python program using the above concatenated string as input and performs the following tasks:

# a. Print the first character of the string.
print(Str_concat[0])

# Print the last character of the string.
print(Str_concat[-1])

#Print the first 5 characters of the string.
print(Str_concat[:5])

#Print the last 11 characters of the string.
print(Str_concat[-11:])

#Print the string in reverse.
print(Str_concat[::-1])

#Use slicing and print the word “Python” from the existing string.
print(Str_concat[24:31])
```

## 3. String Methods

```
from os import replace
# Write a Python program that takes a string, strM = “Python beginner tutorial” and perform the following tasks:

strM = "Python beginner tutorial"
# Convert the sentence to uppercase.
strM = strM.upper()
print(strM)

# Convert the sentence to lowercase.
strM = strM.lower()
print(strM)

# Use Capitalize and return the sentence to the original input form.
strM= strM.capitalize()
print(strM)

# Count the total number of occurrences of character ‘t’ in the string.
Count_letter_t = strM.count('t')
print(Count_letter_t)

# Replace all occurrences of “Python” with “Machine Learning” in the input string strM = “Python beginner tutorial”
print(strM.replace("Python","Machine Learning"))
```

# Tuples


```
# Create 1st tuple with values -> (10, 20, 30), 2nd tuple with values -> (40, 50, 60)
tup1 = (10,20,30)
tup2 = (40,50,60)

# a. Concatenate the two tuples and store it in “t_combine”
t_combine = tup1 + tup2
print(t_combine)

# b. Repeat the elements of “t_combine” 3 times
print(t_combine * 3)

# Access the 3rd element from “t_combine”
print(t_combine[2])

# d. Access the first three elements from “t_combine”
print(t_combine[:3])

# e. Access the last three elements from “t_combine”
print(t_combine[-3:])
```
