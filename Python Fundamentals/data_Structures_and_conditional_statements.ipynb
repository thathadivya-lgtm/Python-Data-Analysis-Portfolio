#List (Creation, Modification and Access)


##1. List Creation


```python
# Create a list named age_list with five integer elements. For eg., [24, 25, 26, 27, 28]
age_list = [23,25,28,32,35]
# Create a list named name_list with five string elements.
name_list =["Divya","Rajesh","Mridhula","Vishwa","Chitra"]

print("age_list: ",age_list)
print("name_list: ",name_list)
```

## 2. List Operations / Modifications

```python
# a. Append the string "Yazhini" to name_list.
name_list.append("Yazhini")
print(name_list)
```

```python
#Insert the element 30 at index 2 in age_list.
age_list.insert(2,30)
print(age_list)
```

```python
#Remove the string "Yazhini" from name_list.
name_list.remove("Yazhini")
print(name_list)
```

```python
#Pop the last element from age_list.
age_list.pop()
print(age_list)
```

```python
# Extend the age_list with additional ages [29, 30, 26].
age_list.extend([29,30,26])
print(age_list)
```

```python
# Sort age_list in descending order.
age_list.sort(reverse=True)
print(age_list)
```

```python
#Find Max age, Min age and sum of all ages from age_list.
Max_age= max(age_list)
print("Max_age: ",Max_age)

Min_age= min(age_list)
print("Min_age: ",Min_age)
Sum_all_ages= sum(age_list)
print("Sum_all_ages: ",Sum_all_ages)
```

## 3. Accessing List Elements


```python
# a. Print the first element of name_list.
print(name_list[0])
```

```python
#b. Print the last element of name_list.
print(name_list[-1])
```

```python
# c. Print the elements from index 2 to index 4 in name_list.
print(name_list[2:5])
```

```python
#Print the elements of name_list in reverse order.
print(name_list[::-1])
```

#Sets (Operations):

```python
#a. Create a set called my_set with following values: ['a','e','i','o','u','a','a','i']
#Analyse the output and provide explanation for the same.
my_set={'a','e','i','o','u','a','a','i'}
print(my_set)
```

###### Set "removes the duplicates" present in the given set.
###### It gives the output in "unordered" way,that means it will not give the output in the same order of input.

```python
# b. Attempt to change the value of my_set[4] = 's'. If code throws an error, provide an explanation.
my_set[4]='s'
```

###### Sets are unordered and do not support indexing.
###### So we cannot access or modify elements using index like my_set[4].

```python
# Create two sets: set1 with values: {1, 3, 5, 7, 9} set2 with values: {2, 3, 5, 8, 10}
Set_1={1, 3, 5, 7, 9}
Set_2={2, 3, 5, 8, 10}
```

```python
from typing_extensions import Union
# Compute and print the union and intersection of set1 and set2.
print("Union of Set_1 and Set_2: ",Set_1.union(Set_2))
print("Intersection of Set_1 and Set_2: ",Set_1.intersection(Set_2))
```

## Performance Category Program

```python
# 1. Prompt user for Input. Score range should be from 0 to 10 (both inclusive).
Score=int(input("Enter your score(0 to 10): "))

#2. Find the performance category based on the input score using following criteria:
# a. Above Average: Score greater than 7
# b. Average: Score between 4 and 7(both inclusive)
# c. Below Average: Score lesser than 4
if(0<=Score<=10):

    if(Score>7):
        print("Above Average: Great job! Your performance is strong—keep pushing to achieve even more!")
    elif(4<=Score<=7):
        print("Average: Good effort! You're progressing well—keep practicing to reach the next level!")
    else:
        print("Below Average: Don't give up! Regular practice will help you get better step by step!")
else:
    print("Please enter a valid score between 0 and 10.")
```
