## **Step 3:** Text Cleaning for Issue Descriptions

```python
# ❖ Clean all issue descriptions by applying:
# ➢ Convert multiple spaces → single space
# ➢ Remove leading/trailing spaces
# ➢ Convert text to lowercase
# ➢ Replace slang/shorthand (example: "ok" → "okay")
# ❖ Hints: Use .replace(), .split(), ' '.join(), .strip(), .lower()
# ➢ Remove punctuation (.,!?-)
for i in range(len(ticket_data['Ticket_No'])):

  ticket = ticket_data['Issue_Description'][i]

  ticket = ticket.replace("!","")
  ticket = ticket.replace(".","")
  ticket = ticket.replace(",","")
  ticket = ticket.strip()
  ticket = ticket.lower()
  ticket = ticket.replace('?','')
  ticket = ticket.replace('-',' ')
  words = ticket.split()
  words = ["okay" if word == "ok" else word for word in words]
  ticket = " ".join(words)

  ticket_data['Issue_Description'][i]=ticket

  print(ticket_data['Issue_Description'][i])
```

##**Step 4:** Keyword-Based Issue Insights


```python
# Create a function def count_tickets_with_word(word):
def count_tickets_with_word(word):
  word=word.lower()
  count=0

  for i in range(len(ticket_data['Ticket_No'])):
    text = ticket_data['Issue_Description'][i]
    if(word in text):
      count+=1
  #Returns how many ticket descriptions contain the given word
  return count

print(f"Number of tickets containing 'poor': {count_tickets_with_word("poor")}")
print(f"Number of tickets containing 'good': {count_tickets_with_word("good")}")
#Case-insensitive search
print(f"Number of tickets containing 'slow': {count_tickets_with_word("sloW")}")
print(f"Number of tickets containing 'excellent': {count_tickets_with_word("Excellent")}")
```

###### **4. Extract Unique Words Used**


```python
# Creating a dictionary to store "word as key" and their "count as values".
word_freq={}
unique_words=set()

for issue in ticket_data['Issue_Description']:
  words=issue.split()
  for word in words:
    if word in word_freq:
      word_freq[word]+=1
    else:
      word_freq[word]=1
print("\033[1mWords and their count in issue descriptions:\033[0m")
pprint(word_freq)
```

```python
# Finding the unique words in the created dictionary and adding it to "unique_words" set
for unique_word,count in word_freq.items():
  if count==1:
    unique_words.add(unique_word)

#Display the count of unique words and sorted unique word lists
print(f"\033[1mCount of all unique words:\033[0m {len(unique_words)}")
```
